```python
top_10_palyed_songs = df_users.groupby("track_id")["playcount"].agg("sum").sort_values(ascending=False).head(10)
```

```python
# users with most different types of songs listedned
most_diverse_listeners = df_users.groupby(["user_id"])["track_id"].agg("count").sort_values(ascending=False).head(5)
```

```python
most_diverse_listeners
```

```python
most_active_users = df_users.groupby("user_id")["playcount"].agg("sum").sort_values(ascending=False).head(5)
most_active_users
```
