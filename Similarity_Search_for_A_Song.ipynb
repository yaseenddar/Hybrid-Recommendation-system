```python
from sklearn.preprocessing import OneHotEncoder,StandardScaler,MinMaxScaler
from category_encoders import CountEncoder
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
```

### for tags it is text based column and we going to use the tfidf and for artist we use the oneHotEncoder cause the column has just one value that is an artitst but tags are so many

```python
#  cols to transform

frequency_encode_cols = ["year"]
ohe_cols = ['artist',"time_signature","key"]
tfidf_col = 'tags'
min_max_scale_cols = ['danceability','energy','speechiness',"acousticness","instrumentalness","liveness","valence"]
standard_scale_cols = ["duration_ms","loudness","tempo"]
```

```python
# transform the data
transformer = ColumnTransformer(transformers=[
    ("frequency_encode",CountEncoder(normalize=True,return_df=True),frequency_encode_cols),
    ("ohe",OneHotEncoder(handle_unknown="ignore"),ohe_cols),
    ("tfidf",TfidfVectorizer(max_features=85),tfidf_col),
    ("standard_scale",StandardScaler(),standard_scale_cols),
    ("min_max_scale",MinMaxScaler(),min_max_scale_cols),
])
```

```python
transformer
```

```python
transformer.fit(df_colab_filtering)
```

```python
transformed_data = transformer.transform(df_colab_filtering)
```

### Calculate Similarity Scores

To calculate similarity, we'll first select a sample song from our `df_colab_filtering` DataFrame. We'll then transform this song's features using the same `transformer` we used for the entire dataset. Finally, we'll compute the cosine similarity between this transformed sample song and all other transformed songs in `transformed_data`.

```python
# Select a sample song (e.g., the first song in the DataFrame)
sample_song_index = 0
sample_song = df_colab_filtering.iloc[[sample_song_index]]

# Transform the sample song using the pre-fitted transformer
input_song_vector = transformer.transform(sample_song)

print(f"Selected song for similarity: \n{sample_song}")
print(f"Shape of the transformed input song vector: {input_song_vector.shape}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculate cosine similarity between the input song vector and all transformed data
similarity_scores = cosine_similarity(input_song_vector, transformed_data)

# The result is a 2D array, so we take the first (and only) row
similarity_scores = similarity_scores.flatten()

# Create a DataFrame to store song names and their similarity scores
similarity_df = pd.DataFrame({
    'song_index': df_colab_filtering.index,
    'artist': df_colab_filtering['artist'],
    'year': df_colab_filtering['year'],
    'similarity': similarity_scores
})

# Sort by similarity in descending order and exclude the song itself
similarity_df = similarity_df.sort_values(by='similarity', ascending=False)

# Display the top 10 most similar songs (excluding the input song itself)
# Since the input song will have a similarity of 1 with itself, it will be the first entry.
print(f"Top 10 songs most similar to song at index {sample_song_index}:\n")
display(similarity_df.iloc[:])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def similarity_scores(df:pd.DataFrame,transformed_data,song_input,k=10):
  # now steps to create the similarity score

  # 1) convert the input into vector form
  input_vector = transformer.transform(song_input)

  # Calculate cosine similarity between the input song vector and all transformed data
  similarity_scores = cosine_similarity(input_vector, transformed_data)

  # top ten songs
  similarity_scores = similarity_scores.flatten()
  similarity_df = pd.DataFrame({
      'song_index': df.index,
      'artist': df['artist'],
      'year': df['year'],
      'spotify_preview_url': df['spotify_preview_url'],
      'similarity': similarity_scores
  })
  return similarity_df.sort_values(by='similarity', ascending=False).reset_index().head(k)
```
